# Test the h5 format model with image 

```python
model = keras.models.load_model('EfficientNetB4_Epoch-48_Val-acc-0.900.h5')
```

```python
##Need this
from io import BytesIO
from urllib import request
from PIL import Image

def download_image(url):
    with request.urlopen(url) as resp:
        buffer = resp.read()
    stream = BytesIO(buffer)
    img = Image.open(stream)
    return img


def prepare_image(img):
    if img.mode != 'RGB':
        img = img.convert('RGB')
    img = img.resize((224, 224), Image.NEAREST)
    return img
```

```python
img = download_image('https://raw.githubusercontent.com/cjj1120/Weather_Image_Classification-Capstone_Project/main/Data/Add/Custom-test-img/rain-test.jpg')
img = prepare_image(img)
img
```

```python
x = np.array(img)
X = np.array([x])
X.shape
```

```python
pred = model.predict(X)
```

## Putting everything together 

```python
##Need this
classes=['fogsmog',
         'hail',
         'lightning',
         'rain',
         'rainbow',
         'rime',
         'sandstorm',
         'snow']

def predict_url(url=''):
    img = download_image(url)
    img = prepare_image(img)
    x = np.array(img)
    X = np.array([x])
    pred = model.predict(X)
    dict_pred = dict(zip(classes, pred[0]))
    return dict_pred
    
```

```python
url ='https://raw.githubusercontent.com/cjj1120/Weather_Image_Classification-Capstone_Project/main/Data/Add/Custom-test-img/rainbow-test.jpg'
pred = predict_url(url=url)
pred
```

# Convert model to tflite

```python
# Convert model to tflite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

with open('weather-model.tflite', 'wb') as f_out:
    f_out.write(tflite_model)
```
